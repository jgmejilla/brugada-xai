```python
from datetime import date

today = date.today()
DATE = f'{today.month}-{today.day}-{today.year}'
DATE
```

```python
from pathlib import Path
import numpy as np

BASE_PATH = Path.cwd().parent / 'samples'
POSITIVE_PATH = BASE_PATH / 'positive'
NEGATIVE_PATH = BASE_PATH / 'negative'
POSITIVE_LIFECARE_PATH = BASE_PATH / 'positive-lifecare'

# save to today's date
RESULTS_PATH = Path.cwd().parent / 'results' / DATE / 'ecg_visualization'
RESULTS_PATH.mkdir(parents=True, exist_ok=True)

paths = [POSITIVE_PATH, NEGATIVE_PATH, POSITIVE_LIFECARE_PATH, RESULTS_PATH]

assert all([path.exists() for path in paths])

for path in RESULTS_PATH.rglob('*'):
    if path.is_file(): 
        path.unlink()
```

```python
positives = [filepath for filepath in POSITIVE_PATH.iterdir() if filepath.suffix == '.npy']
negatives = [filepath for filepath in NEGATIVE_PATH.iterdir() if filepath.suffix == '.npy']
positives_lifecare = [filepath for filepath in POSITIVE_LIFECARE_PATH.iterdir() if filepath.suffix == '.npy']

f'{len(positives)=}, {len(negatives)=}, {len(positives_lifecare)=}'
```

```python
import importlib
import feature_extraction as fe
import matplotlib.pyplot as plt

importlib.reload(fe)

def run_samples(paths):
    for path in paths:
        signal = np.load(path, allow_pickle=True)

        name = path.name
        stem = path.stem
        image_filename = f'{stem}.png'

        assert RESULTS_PATH.exists(), "results path doesn't exist"

        print(f'processing {name}')   
        try:        
            fe.display_ecg(signal, title=name, dpi=200, save_path=(RESULTS_PATH / image_filename), landmarks=False)
        except Exception as e:
            print(f'error processing {name}: {e}')
            plt.close()
            continue
```

```python
run_samples(negatives[:10])
```
